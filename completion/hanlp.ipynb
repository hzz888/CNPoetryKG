```python
import json 
import xlrd
import xlwt
import json
```

```python
poemData = xlrd.open_workbook('final.xls')
table = poemData.sheets()[0]
```

```python
authorLocaList = []
poemLocaList = []

for i in range(1, table.nrows):
    row = table.row_values(i)

    authorIntro = row[3]
    poemInfo = row [4] + row[5] +row[6] + row[7]

    authorSens = split_sent(authorIntro)
    poemSens= split_sent(poemInfo)

    authorLocas = []
    poemLocas = []

    for au in authorSens:
        pos_result_a = HanLP(au, tasks='pos/pku').to_dict()
        for index1,type_entity_1 in enumerate(pos_result_a['pos/pku']):
          if type_entity_1 == 'ns':
            content = pos_result_a['tok/fine'][index1]
            if content != '中国':
                authorLocas.append(content)
            else:
              pass
          else:
            pass
    authorLocas = list(set(authorLocas))
    print('已完成第' + str(i) + '位诗人的相关地点提取 ' + ','.join(authorLocas))
    authorLocaList.append(authorLocas)
    
    for po in poemSens:
        pos_result_p = HanLP(po, tasks='pos/pku').to_dict()
        for index2,type_entity_2 in enumerate(pos_result_p['pos/pku']):
          if type_entity_2 == 'ns':
            content = pos_result_p['tok/fine'][index2]
            if content != '中国':
                poemLocas.append(content)
            else:
              pass
          else:
            pass  
    poemLocas=list(set(poemLocas))
    print('已完成第' + str(i) + '首诗词的相关地点提取 ' + ','.join(poemLocas))
    poemLocaList.append(poemLocas)

locaFile = xlwt.Workbook()
locaSheet = locaFile.add_sheet('sheet1')
locaSheet.write(0, 0, '诗人相关地点')
locaSheet.write(0,1,'诗词相关地点')

j = 0
for i in range(1, len(poemLocaList)+1):
    locaSheet.write(i, 0, ','.join(authorLocaList[j]))
    locaSheet.write(i, 1, ','.join(poemLocaList[j]))
    j += 1

locaFile.save('locations.xls')
print('诗人地点信息补全完成')
```
